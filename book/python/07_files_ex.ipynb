```python
input_file = "coordinates.txt"
output_file = "output_coordinates.txt"
```

```python
try:
    with open (input_file, "r") as file:
        coordinates = file.readlines()
        #print(f'file has been sucessully read {file}')
        print(coordinates)

    with open ( output_file, "w") as outfile:
        for line in coordinates:
            lat,lon = line.strip().split(",")
            outfile.write(f' {lat}, {lon}\n')
    print(f'files have been printe to file {outfile}')

except FileNotFoundError:
    print(f'the file {input_file} has not been found')
    print(f'upload a new file or fix it')



```
